```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IrisMissingData.csv')
df
```

```python
nan_exist = df.isna().any().any()
print(f"Are there any NaN values in the dataset? {nan_exist}")
```

```python
nan_counts = df.isna().sum()
print(nan_counts)
```

```python
# Look for any rows that have NA
rows_missing_values = df.isna().any(axis=1)
df[rows_missing_values]
```

```python
group_means = df.groupby('class').transform('mean')
df_filled = df.fillna(group_means)
df_filled
```
